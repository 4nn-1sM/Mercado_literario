```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
df_amazon = pd.read_csv("../data/bestsellers_with_categories_2022_03_27.csv")
```

```python
ranking = list(range(1,df_amazon.shape[0]+1))
#df_amazon["Ranking"] = ranking
df_amazon["title_raw"] = df_amazon["Name"].str.strip().str.lower().str.replace(":","").str.replace(",","").str.replace(".","")
```

```python
years = df_amazon["Year"].unique()
```

```python
# creación de una columna por cada año en la que se indica la posición del libro ese año en el top 50. Quizás sería mejor separar el dataframe en años,
#  hacer este ranking y luego hacer un merge que incluya en la lista generica solo ese valor, ya que los títulos están duplicados por año. Otro problema 
# es que no pone en qué formato se están vendiendo los libros. por eso hay más de un libro duplicado por año, porque depnde del formato del libro. creo que 
# hay tanto ebook como tapa blanda y tapa dura... minimo

df_amazon_ranking = df_amazon.copy()
for year in years:
    print(year)
    df_year = df_amazon[df_amazon["Year"] == year].copy()
    df_year["top_50_" + str(year)] = range(1, len(df_year) + 1)
    df_amazon_ranking = pd.merge(df_amazon_ranking,df_year, how="left")
```
