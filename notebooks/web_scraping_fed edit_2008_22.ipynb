```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
# import urllib3 # urllib3 es un cliente HTTP potente y fácil de usar para Python.
import re # Expresiones regulares 
import time
import pandas as pd
```

```python
service = Service(executable_path='../utils/chromedriver.exe')
options = webdriver.ChromeOptions()

driver = webdriver.Chrome(service=service, options=options)
```

```python
# primero extraemos los datos desde 2020 a 2022
driver.get("https://www.federacioneditores.org/datos-estadisticos-2020.php")
time.sleep(5)
```

```python
element_by_xpath = driver.find_element(By.XPATH, '/html/body/div[2]/section[3]/div/div/div/div/div/div/table')
time.sleep(8)
element_by_xpath.text
filas = element_by_xpath.text.split("\n")
filas
#tabla = eval(string)
#tabla
```

```python
driver.quit()
time.sleep(9)
```

```python
columnas = ("datos_estadisticos " + filas[0]).split()
columnas
```

```python
filas = filas[1:]
filas
```

```python
data = []

for linea in filas:
    data.append(linea.rsplit(maxsplit=3))
data
```

```python
tabla_completa_2020_22 = pd.DataFrame(data, columns=columnas)
tabla_completa_2020_22
```

```python
datos_absolutos_2020_22 = tabla_completa_2020_22[:12]
```

```python
tabla_completa_2020_22.columns
```

```python
distrib_cifra_neg_genero_2020_22 = tabla_completa_2020_22[13:]
distrib_cifra_neg_genero_2020_22 = distrib_cifra_neg_genero_2020_22.rename(columns={"datos_estadisticos": "distribucion_cifra_negocio_por_genero"})
distrib_cifra_neg_genero_2020_22
```

```python
# A partir de aquí obtenenos los datos de 2008 a 2019
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.federacioneditores.org/datos-estadisticos.php")
time.sleep(7)
```

```python
element_by_xpath = driver.find_element(By.XPATH, '/html/body/div[2]/section[3]/div/div/div/div/div/div/table')
time.sleep(10)
```

```python
element_by_xpath.text
filas = element_by_xpath.text.split("\n")
filas
#tabla = eval(string)
#tabla
```

```python
driver.quit()
```

```python
columnas = ("datos_estadisticos " + filas[0]).split()
columnas
```

```python
filas = filas[1:]
filas
```

```python
data = []

for linea in filas:
    data.append(linea.rsplit(maxsplit=12))
data
```

```python
tabla_completa_2008_19 = pd.DataFrame(data, columns=columnas)
tabla_completa_2008_19
```

```python
datos_absolutos_2008_19 = tabla_completa_2008_19[:14]
datos_absolutos_2008_19
```

```python
tabla_completa_2008_19.columns
```

```python
distrib_cifra_neg_genero_2008_19 = tabla_completa_2008_19[16:]
distrib_cifra_neg_genero_2008_19 = distrib_cifra_neg_genero_2008_19.rename(columns={"datos_estadisticos": "distribucion_cifra_negocio_por_genero"})
distrib_cifra_neg_genero_2008_19
```

```python
# con todos los datos extraidos los concatenamos con merge en un unico dataframe de cada tipo de tabla
datos_absolutos = pd.merge(datos_absolutos_2008_19,datos_absolutos_2020_22, how= "outer")
datos_absolutos
```

```python
distrib_cifra_neg_generos = pd.merge(distrib_cifra_neg_genero_2008_19,distrib_cifra_neg_genero_2020_22, how= "outer")
distrib_cifra_neg_generos
```

```python
datos_absolutos.to_csv("../data/abs_sect_editorial_esp.csv")
distrib_cifra_neg_generos.to_csv("../data/dist_cif_neg_genero_sect_editorial_esp.csv")
```
